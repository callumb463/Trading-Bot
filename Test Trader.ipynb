```python
import yfinance as yf
import pandas as pd
import numpy as np
import math
import backtest as bt
import chart_studio.plotly as py
import plotly.express as px
import cufflinks as cf
import plotly.graph_objects as go

from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
%matplotlib inline
init_notebook_mode(connected=True)
cf.go_offline
```

```python
tickers = ['AAPL']
timeframe = '10y'

df=yf.download(tickers, period=timeframe,rounding=True)
```

```python
def close(df):
    return df[['Close']]
```

```python

class Test(bt.Strategy):
    def __init__(self, data, cash):
        super().__init__(data, cash)
        self.indicator(close, data)
        self.timer = 0
    
    def run(self, date_index, row):
        super().run(date_index,row)
        #if (math.floor(row['Close'])%5 == 0):
        #    self.buy('AAPL',portfolio_perc=0.1)
        #if (math.floor(row['Close'])%4 == 0):
        #    self.sell('AAPL')
        if self.timer == 0:
            self.buy('AAPL',buy_price=1)
        elif self.timer == 1501:
            self.buy('AAPL',buy_price=5.4)
        elif self.timer == 1500 or self.timer == 2500:
            self.sell('AAPL')
        self.timer += 1
            

```

```python
test = Test(df, 10_000)
```

```python
testrun = bt.Backtest(df, test)
testrun.run_backtest()
```

```python
testrun.visualize()
```

```python
profit = 0
for trade in test.all_trades:
    profit += trade.price-trade.buy_price
profit
```
